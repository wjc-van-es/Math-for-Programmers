```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f_square(x):
    return x * x
```

```python
input = range(0,20)
output = []
for x in range(0,20):    
    output.append(f_square(x))
    
print(output)
```

```python
xs = np.linspace(0,21,370)
ys = [f_square(x) for x in xs]
plt.plot(xs,ys, color='C1')
plt.scatter(input, output)
plt.xlabel('x')
plt.ylabel('y = x^2')
plt.savefig('x^2.svg')
```
